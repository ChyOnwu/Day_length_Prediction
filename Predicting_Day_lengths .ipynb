```python
#Task 2
#1)

import numpy as np 
import matplotlib.pyplot as plt


x = np.array([[0],[3],[6]])
X = np.array([[1,0],[1,3],[1,6]])
y = np.array([[1],[4],[5]])

print('x:',('\n'),x)
print('X:',('\n'),X)
print('y:',('\n'),y)


X_trans = (np.transpose(X))@X
print('Transpose of X:',('\n'),X_trans)

Inv_X_trans = np.linalg.inv(X_trans)
print('Inverse of X transpose:',('\n'),Inv_X_trans)

y_trans = (np.transpose(X))@y
print('Inverse@y:',('\n'),y_trans)

beta = Inv_X_trans@y_trans
print('Beta:',('\n'),beta)
X_beta = X@beta
print('Fitted/Predicted Value:',('\n'),X_beta)


plt.scatter(x, y)
plt.xlim([-1, 7])
plt.ylim([0, 6])
plt.plot(x, X_beta)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

```

```python
#“residuals vs fitted” graph
residuals = y-X_beta
print('Residual Value:',('\n'),residuals)


plt.scatter(X_beta,residuals)
plt.xlabel('Fitted')
plt.ylabel('Residuals')
plt.grid()
plt.show()
```

```python
#Coefficient of Determination (R squared value)
actual = y
predict = X_beta

print('Actual Values:',('\n'),y)
print('Residual Value:',('\n'),X_beta)
 
corr_matrix = np.corrcoef(np.hstack(actual),np.hstack(predict))
print('Correlation Matrix:',('\n'),corr_matrix)
corr = corr_matrix[0,1]
R_sq = corr**2
 
print('R squared value:',R_sq)
```

```python
#Compare results on the tiny example given above with the output from numpy.polyfit().
plt.scatter(x, y)
#obtain a (slope) and b(intercept) of linear regression line
b,a = np.polyfit(np.hstack(x),np.hstack(y), 1)
#add linear regression line to scatterplot 
plt.plot(x, a+b*x)
plt.xlim([-1, 7])
plt.ylim([0, 6])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
